```python
import pandas as pd
from pycaret.time_series import *
```

```python
df = pd.read_excel('../Input/DadosCompeticao.xlsx')
```

```python
df.index = pd.RangeIndex(start=0, stop=len(df), step=1)
```

```python
forecast_results = {}
scores = {}

for col in df.columns:
    print(f"\n==== Treinando {col} ====")

    ts = df[col]  # SÃ©rie univariada
    exp = setup(
        data=ts,
        fold=3,           # split em 3 kfolds para cross val
        fh=12,            # forecast horizon
        session_id=42,
        verbose=False,
    )

    best_model = compare_models(sort='RMSE')


    result_metrics = pull() # Comparamos os scores
    rmse_score = result_metrics.iloc[0]['RMSE']

    final_model = finalize_model(best_model)
    future_forecast = predict_model(final_model, fh=12)

    forecast_results[col] = future_forecast
    scores[col] = rmse_score
```

```python
first_series = list(forecast_results.keys())[0]
index = forecast_results[first_series].index

forecast_df = pd.DataFrame(index=index)

for series_name, forecast in forecast_results.items():
    forecast_df[series_name] = forecast['y_pred']
```

```python
scores_df = pd.DataFrame.from_dict(scores, orient='index', columns=['RMSE'])
```

```python
forecast_df.to_excel('../Output/v1/previsoes_automl.xlsx', index=False)
scores_df.to_excel('../Output/v1/scores_automl.xlsx', index=True)
```

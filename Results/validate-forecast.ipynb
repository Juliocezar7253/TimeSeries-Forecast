```python
import pandas as pd
import plotly.express as px
```

### Dataset de input

```python
df = pd.read_excel('../Input/DadosCompeticao.xlsx')
```

```python
df_ = df.reset_index().melt(id_vars='index', var_name='serie', value_name='value')
df_.rename(columns={'index': 'time'}, inplace=True)
```

### Dados de Previsão (Modelo - Xgboost)

```python
forecast = pd.read_excel('../Output/v5/previsoes_xgboost.xlsx')
```

```python
forecast_ = forecast.reset_index().melt(id_vars='index', var_name='serie', value_name='value')
forecast_.rename(columns={'index': 'time'}, inplace=True)
```

```python
max_real_tempo = df_['time'].max()
forecast_['time'] = forecast_['time'] + max_real_tempo + 1

# Segregar o tipe em real e previsto
df_['type'] = 0
forecast_['type'] = 1
combined_df = pd.concat([df_, forecast_])

fig = px.line(
    combined_df,
    x='time',
    y='value',
    color='serie',
    line_dash='type',
    title='Séries: Real + Previsto (fh: 12)',
    template='ggplot2'
)

fig.update_layout(
    legend_title_text='Series',
    legend=dict(
        orientation='v',
        y=1,
        x=1.05,
        xanchor='left',
        yanchor='top'
    ),
    height=500,
    width=1000
)

fig.write_image("forecast_plot.png")
fig.show()
```
